```python
# credit 和 movie 两个数据合并
data= pd.merge(credit,movie,left_on='movie_id',right_on='id')
data.head(3)
print(data.shape)
print('credit shape',credit.shape)
print('movie shape',movie.shape)
```

```python
# title, id 两个表数据一致没必要都保留，删除一个
data.drop(columns=['title_x','id'],inplace=True)
data.shape
```

```python
data.head(3)
```

```python
# release date 转化为时间格式
data['release_date'] = pd.to_datetime(data['release_date'])
print(data['release_date'].describe())
```

```python
# 只取出日期
data['date']= data['release_date'].dt.date
data['date'].describe()
```

```python
# 部分列的数据是json 格式
cols= ['cast','crew','keywords','genres','production_countries','production_companies','spoken_languages']

print(type(data['production_countries'][0]))  #<class 'str'>
print(data['production_countries'][0])

# 原始是字符串格式
# 转化为json
import json
for col in cols:
    data[col]=data[col].apply(json.loads)
    
print(type(data['production_countries'][0]))
print(data['production_countries'][0])
```

```python
# 从crew 里面拿出导演
data['crew'][0]
print(data['crew'][0])
```

```python
# 导演的 job value 为director

def get_director(crew_data):
#     for item in crew_data:
#         if item['job']== 'Director':
#             return itme['name']
    # 用列表推到式为
    if crew_data:
        return [item['name'] for item in crew_data if item['job']=='Director']
    else:
        np.nan

data['director'] = data['crew'].apply(get_director)
```

```python
print(data.isnull().sum())
print(r'==='*3)
print(data.shape)
print(r'==='*3)
print(data['director'].describe())
```

```python
# 不存在导演的个数比列
print('director not exist number',data['director'].isnull().sum())
print('director not exist percentage: ',data['director'].isnull().sum()/len(data['director']) *100,'%')

# 查看某个电影的导演是否有大于两个的
def director_gt2(director):
    if director and len(director)>1:
        return True
    else:
        return False

data['director'].apply(director_gt2).sum()
```

```python
# cols= ['cast','crew','keywords','genres','production_countries','production_companies','spoken_languages']

data['production_countries'][3]
# 从production_countries 里面取出国家name

def get_country_name(production_countries):
    return [item['name'] if production_countries else np.nan for item in production_countries]

data['country']=data['production_countries'].apply(get_country_name)

print('country is null number',data['country'].isnull().sum())
print('country number great than 2: ', data['country'].apply(lambda x: True if len(x)>1 else False ).sum())
print('country great than 2: ', data['country'].apply(lambda x: x if len(x)>1 else False ))
```

```python
# 从cast 里面提取 actor
data['cast'][0]
# 暂时放弃
```

```python
# 提取keywords
print(data['keywords'][0])

# 提取keyword 用| 分割
def split_keyword(keywords):
    keywords= [keyword['name'] if keywords else np.nan for keyword in keywords]
    return '|'.join(keywords)

data['keywords_name']= data['keywords'].apply(split_keyword)
print(data['keywords_name'].isnull().sum())
```

```python
set_keywords= set()
print(data['keywords_name'].str.split('|').values[0])
print('==='*9)
print(data['keywords_name'].str.split('|').values)
for keywords in data['keywords_name'].str.split('|').values:
    if isinstance(keywords,float):
        continue # only happen if liste_keywords = NaN
    set_keywords = set_keywords.union(keywords)

# remove null chain entry
set_keywords.remove('')
set_keywords
```

```python
# 制作keyword 云图
# 把所有的word 拿出用作为字符串的一部分

keywords= list()
for keyword in data['keywords_name'].str.split('|').values:
    keywords.append(keyword)
    
# 将两层列表平铺开来 itertools

from itertools import chain
keywords=list(chain(*keywords))
```

```python
# 将list 转化为字符串
str=' '.join(keywords)
```

```python
# 制作keyword 云图
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud= WordCloud(background_color='Black',max_words=5,max_font_size=1000)

wordcloud.generate(str)
# plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
# 那个国家的电影最多
country_list=data['country'].values.tolist()
# 将多层list 平铺开来
country_list = list(chain(*country_list))
country_set=set(country_list)
# 统计list里面的个数
from collections import Counter
counter = dict(Counter(country_list))
# 安装value 次数排序
sorted_counter= dict(sorted(counter.items(), key=lambda x: x[1], reverse=True))
# 
x = list(sorted_counter.keys())[:10]
y = list(sorted_counter.values())[:10]
print(x)
print(y)
plt.figure(figsize=(12,8))
plt.pie(y,labels=x,autopct='%1.2f%%')
plt.show()
```

```python
# 分析电影类型的变化
data['genres'][2009]
# 用apply 方法把每个genre 里面的name 拿出来
def get_genres(genre_list):
    genres=[genre['name'] for genre in genre_list]
    return genres

data['genres'] = data['genres'].apply(get_genres)
```

```python
# 统计每种genres 的数量
# 取出这一列所有的值
genres = data['genres'].values.tolist()

# 把每个list 里面的value 平铺开来
genres_list = list(chain(*genres))
print(len(genres_list))
print(len(set(genres_list)))
```

```python
# 计算genres_list 里面每个genre 的数量
counter = Counter(genres_list)
print(counter)
```
